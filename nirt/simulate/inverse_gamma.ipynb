```python
# Inverse Gamma distribution check.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import invgamma
```

```python
for alpha, beta in ((1,1), (2,1), (3,1), (3, 0.5)):
    rv = invgamma(a=alpha, scale=beta)
    x = np.linspace(0, np.minimum(rv.dist.b, 3), 1000)
    h = plt.plot(x, rv.pdf(x))    
plt.grid(True)
plt.legend(['a=1,b=1','a=2,b=1', 'a=3,b=1', 'a=3,b=0.5'])
plt.title('Probability density function')
```

```python
for alpha, beta in ((1,1), (2,1), (3,1), (3, 0.5)):
    rv = invgamma(a=alpha, scale=beta)
    x = np.linspace(0, np.minimum(rv.dist.b, 3), 1000)
    h = plt.plot(x, rv.cdf(x))    
plt.grid(True)
plt.legend(['a=1,b=1','a=2,b=1', 'a=3,b=1', 'a=3,b=0.5'])
plt.title('Cumulative distribution function')
```

```python
# Draw samples from the distribution.
rv = invgamma(a=3, scale=1)
r = rv.rvs(10000)
plt.hist(r, bins=500, density=True)
plt.xlim([0,3])
x = np.linspace(0, 3, 1000)
h = plt.plot(x, rv.pdf(x), 'r-') 
```
